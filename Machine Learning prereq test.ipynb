# Question2

```python
def matrix_multiply(a, b):
    result = []
    for i in range(len(a)):
        row = []
        for j in range(len(b[0])):
            row.append(sum(a[i][k] * b[k][j] for k in range(len(a[0]))))
        result.append(row)
    return result

# Example matrices
A = [[1, 2, 3], [4, 5, 6]]
B = [[7, 8], [9, 10], [11, 12]]

# Perform matrix multiplication using list of lists
result_list_of_lists = matrix_multiply(A, B)
print("Matrix multiplication using list of lists:")
print(result_list_of_lists)

```

```python
import numpy as np


A_np = np.array(A)
B_np = np.array(B)

# Performing matrix multiplication using NumPy
result_numpy = np.dot(A_np, B_np)
print("\nMatrix multiplication using NumPy:")
print(result_numpy)
```

```python
import timeit

# Timing list of lists multiplication
time_list_of_lists = timeit.timeit("matrix_multiply(A, B)", globals=globals(), number=1000)

# Timing NumPy multiplication
time_numpy = timeit.timeit("np.dot(A_np, B_np)", globals=globals(), number=1000)

print(f"\nTime for list of lists multiplication: {time_list_of_lists:.6f} seconds")
print(f"Time for NumPy multiplication: {time_numpy:.6f} seconds")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Function to normalize y-values
def normalize_function(y_values):
    scaler = MinMaxScaler(feature_range=(0, 1))
    normalized_values = scaler.fit_transform(y_values.reshape(-1, 1)).flatten()
    return normalized_values

x_values = np.arange(0.5, 100.5, 0.5)

y1 = x_values
y2 = x_values**2
y3 = x_values**3 / 100
y4 = np.sin(x_values)
y5 = np.sin(x_values) / x_values
y6 = np.log(x_values)
y7 = np.exp(x_values)

# Normalize y-values
y1_normalized = normalize_function(y1)
y2_normalized = normalize_function(y2)
y3_normalized = normalize_function(y3)
y4_normalized = normalize_function(y4)
y5_normalized = normalize_function(y5)
y6_normalized = normalize_function(y6)
y7_normalized = normalize_function(y7)

# Create a single set of axes
fig, ax = plt.subplots(figsize=(10, 6))

# Plot normalized curves on the same set of axes
ax.plot(x_values, y1_normalized, label=r'$y = x$')
ax.plot(x_values, y2_normalized, label=r'$y = x^2$')
ax.plot(x_values, y3_normalized, label=r'$y = \frac{x^3}{100}$')
ax.plot(x_values, y4_normalized, label=r'$y = \sin(x)$')
ax.plot(x_values, y5_normalized, label=r'$y = \frac{\sin(x)}{x}$')
ax.plot(x_values, y6_normalized, label=r'$y = \log(x)$')
ax.plot(x_values, y7_normalized, label=r'$y = e^x$')

# Set labels and legend
ax.set_xlabel('$x$')
ax.set_ylabel('Normalized $y$')
ax.legend()

# Show the merged plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd

random_matrix = np.random.uniform(1, 2, size=(20, 5))

df = pd.DataFrame(random_matrix, columns=['a', 'b', 'c', 'd', 'e'])

highest_std_column = df.std(axis=0).idxmax()
lowest_mean_row = df.mean(axis=1).idxmin()

print(df)
print("\n")
print("Column with the highest standard deviation:", highest_std_column)
print("Row with the lowest mean:", lowest_mean_row)

```

# Question 11

```python
import numpy as np
import pandas as pd

random_matrix = np.random.uniform(1, 2, size=(20, 5))

df = pd.DataFrame(random_matrix, columns=['a', 'b', 'c', 'd', 'e'])

df['f'] = df[['a', 'b', 'c', 'd', 'e']].sum(axis=1)

df['g'] = np.where(df['f'] < 8, 'LT8', 'GT8')

lt8_count = (df['g'] == 'LT8').sum()

std_lt8 = df.loc[df['g'] == 'LT8', 'f'].std()
std_gt8 = df.loc[df['g'] == 'GT8', 'f'].std()

print(df)
print("\nNumber of rows where 'g' is 'LT8':", lt8_count)
print("Standard deviation of 'f' where 'g' is 'LT8':", std_lt8)
print("Standard deviation of 'f' where 'g' is 'GT8':", std_gt8)
```
